```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: 2 * x**2 + 1

xs = np.linspace(-10, 10, 100)
xd = np.random.rand(1,10,100)
yd = f(xd) + 0.1*np.random.rand(1,10,100) 

plt.title("quadratic function graph")
plt.scatter(xd, yd, s=1)
plt.plot(xs, f(xs), label=f"y = 2x^2 + 1")
plt.legend()
plt.show()
```
